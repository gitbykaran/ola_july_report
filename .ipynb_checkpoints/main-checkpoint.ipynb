### Analysis (SQL Queries)

<br>

1. Retrieve all successful bookings

```python
%%sql 
SELECT * FROM ola_july 
WHERE Booking_Status = 'Success';
```

<br>

2. What is the average ride distance for each vehicle type?

```python
%%sql
SELECT
Vehicle_Type,
AVG(Ride_Distance) AS avg_dist
FROM ola_july
GROUP BY Vehicle_Type;
```

<br>

3. What is the total number of cancelled rides by customers?

```python
%%sql 
SELECT 
COUNT(Canceled_Rides_by_Customer) cnt
FROM 
ola_july;
```

<br>

4. What are the top 5 customers who booked the highest number of rides?

```python
%%sql
SELECT 
Customer_ID,
COUNT(Booking_ID)
FROM ola_july
GROUP BY Customer_ID
ORDER BY COUNT(Booking_ID) DESC
LIMIT 5 ;
```

<br>

5. What is the number of rides cancelled by drivers due to personal and car-related issues?

```python
%%sql
SELECT 
COUNT(*) AS cnt
FROM ola_july
WHERE Canceled_Rides_by_Driver = 'Personal & Car related issue';
```

<br>

6. What is the maximum and minimum driver ratings for Prime Sedan bookings?

```python
%%sql
SELECT 
MAX(Driver_Ratings) AS max,
MIN(Driver_Ratings) AS min
FROM ola_july
WHERE Vehicle_Type = 'Prime Sedan';
```

<br>

7. All rides where payment was made using UPI.

```python
%%sql
SELECT *
FROM ola_july
WHERE Payment_Method = 'UPI' ;
```

<br>

8. What is the average customer rating per vehicle type?

```python
%%sql
SELECT 
Vehicle_Type,
ROUND(AVG(Customer_Rating),2) AS avg_rating
FROM ola_july
GROUP BY Vehicle_Type;
```

<br>

9. What is the total booking value of rides completed successfully?

```python
%%sql 
SELECT 
SUM(Booking_Value) AS total_booking_val
FROM ola_july 
WHERE Booking_Status = 'Success';
```

<br>

10. All incomplete rides along with the reason.

```python
%%sql 
SELECT 
Customer_ID,
Booking_ID,
Incomplete_Rides_Reason
FROM ola_july
WHERE Incomplete_Rides = 'Yes';
```
